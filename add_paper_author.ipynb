```python
def special_case(author, author_id):
    #if author == "Dawn Song":
    #   return author, "dawnsong@cs.berkeley.edu"
    if author == "Richard Socher":
        return author, "rsocher@salesforce.com"
    if author == "Aaron Courville":
        return author, "aaron.courville@umontreal.ca"
    if "nandkumar" in author:
        return author, "anima@caltech.edu"
    #if author == "Bo Li":
    #    return author, "bli@cshl.edu"
    return author, author_id
```

```python
for key, values in author_dic.items():
    if values[2] > 0:
        values[2] /= values[1]
    if values[4] > 0:
        values[4] /= values[3]
    if values[6] > 0:
        values[6] /= values[5]
    if values[8] > 0:
        values[8] /= values[7]
    if values[10] > 0:
        values[10] /= values[9]
```

```python
df_list = []
for key, values in author_dic.items():
    for inst in values[0]:
        if 'gmail' not in inst:
            insti = inst
            break
    else:
        insti = 'unknown@unknown.unknown'
    df_temp = pd.DataFrame({'id':insti,'avg_nb_cohauthor':values[-1],'nb_paper':values[1],'avg_note':values[2],'nb_reject':values[3],'avg_reject':values[4],'nb_workshop':values[5],'avg_workshop':values[6],'nb_poster':values[7],'avg_poster':values[8],'nb_oral':values[9],'avg_oral':values[10]} ,index=[key])
    df_list.append(df_temp)
df_author = pd.concat(df_list)
df_author.avg_nb_cohauthor = df_author.avg_nb_cohauthor / df_author.nb_paper
```

```python
df_author.sort_values(['nb_paper'],ascending=False)
```

```python
df_author['institution'] = df_author.id.apply(lambda x: x.split('@')[-1])
df_author['institution_country'] = df_author.id.apply(lambda x: x.split('.')[-1])
```

```python
df_author['acceptance_rate'] =  (df_author.nb_poster + df_author.nb_oral) / df_author.nb_paper
```
